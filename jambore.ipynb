```python
import matplotlib.pyplot as plt
import numpy as np

train_sizes = np.linspace(0.1, 1.0, 10)  # Vary training set size
from sklearn.model_selection import learning_curve
from sklearn.linear_model import LinearRegression  # Change this for different models

model = LinearRegression()
train_sizes, train_scores, test_scores = learning_curve(model, X, y, train_sizes=train_sizes, cv=5)

train_mean = np.mean(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)

plt.figure(figsize=(8,5))
plt.plot(train_sizes, train_mean, label="Training Score", marker='o', color="blue")
plt.plot(train_sizes, test_mean, label="Validation Score", marker='o', color="red")

plt.xlabel("Training Set Size")
plt.ylabel("Score (Accuracy/R)")
plt.title("Learning Curve")
plt.legend()
plt.show()
```

 Bias-Variance Trade-off Explained


Factor	               High Bias (Underfitting) 	             High Variance (Overfitting) 


Model Complexity	  Too simple (e.g., Linear Regression)	     Too complex (e.g., Deep Neural Networks)


Training Error	                          High	                     Very low


Test Error                           	High	                          High (due to overfitting)


Flexibility	                         Low                                    	High

Generalization	                          Poor	                                      poor
